```python
def validate_inputs(audience, tone):

    valid_audience = [
        "beginner",
        "intermediate",
        "expert"
    ]

    valid_tone = [
        "formal",
        "casual",
        "fun"
    ]

    if audience not in valid_audience:
        print("Invalid audience. Defaulting to beginner.")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone. Defaulting to formal.")
        tone = "formal"

    return audience, tone
```

```python
audience, tone = validate_inputs(
    "kids",
    "friendly"
)

print(audience)
print(tone)
```

```python
styles = {

    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style."
    ),

    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Create interview questions on {topic} for {audience} in a {tone} tone."
    ),

    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style."
    )
}
```

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style not in styles:
        raise ValueError("Invalid Style")

    return styles[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(
    generate_prompt(
        "Neural Networks",
        "beginner",
        "fun",
        "storytelling"
    )
)
```

```python
reuse_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple language."
)
```

```python
topics = [
    "AI",
    "Python",
    "SQL",
    "Data Science",
    "Cloud Computing"
]

for topic in topics:
    print(reuse_template.format(topic=topic))
```

```python
print("Pipeline Flow")

topic = "Machine Learning"

audience = "beginner"

tone = "formal"

style = "teaching"

audience, tone = validate_inputs(audience, tone)

result = generate_prompt(
    topic,
    audience,
    tone,
    style
)

print(result)
```
